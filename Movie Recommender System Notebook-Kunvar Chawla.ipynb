## Have to convert the lists in the columns Genre,Keyword and Cast from strings to list

```python
import ast
```

```python
columns_tolist = ['Generes', 'Keywords', 'Top 5 Casts']
```

```python
movies_imdb[columns_tolist]
```

```python
for i in columns_tolist:
    movies_imdb[i] = movies_imdb[i].apply(ast.literal_eval)
```

```python
movies_imdb[columns_tolist]
```

```python
movies_imdb.info()
```

```python
movies_imdb[movies_imdb.Overview.isnull()]
```

```python
##dropping movies where overview not present
movies_imdb.dropna(subset=['Overview'], inplace=True)
```

```python
movies_imdb.shape
```

```python
movies_imdb
```

```python
## Removing whitespaces from Genre, keywords, cast, director, writer
```

```python
def remove_spaces(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
list_columns = ['Generes', 'Keywords','Top 5 Casts']
```

```python
for i in list_columns:
    movies_imdb[i] = movies_imdb[i].apply(remove_spaces)
```

```python
movies_imdb[list_columns]
```

```python
## Now for director and writer
movies_imdb['Director'] = movies_imdb['Director'].apply(lambda x:x.replace(" ",""))
movies_imdb['Writer'] = movies_imdb['Writer'].apply(lambda x:x.replace(" ",""))
```

```python
#Turning overview paragraph into list

movies_imdb['Overview'] = movies_imdb.Overview.apply(lambda x: x.split())
```

```python
#Turning Director and Writer into lists
movies_imdb['Director'] = movies_imdb['Director'].apply(lambda x:[x])
movies_imdb['Writer'] = movies_imdb['Writer'].apply(lambda x:[x])
```

```python
##Creating tags column
```

```python
movies_imdb['tags'] = movies_imdb['Generes'] + movies_imdb['Overview'] + movies_imdb['Keywords'] + movies_imdb['Director'] + movies_imdb['Top 5 Casts'] + movies_imdb['Writer']
```

```python
def to_lower_case(L):
    L1 = []
    for i in L:
        L1.append(str.lower(i))
    return L1
```

```python
movies_imdb['tags'] = movies_imdb['tags'].apply(to_lower_case)
```

```python
movies_new = movies_imdb.drop(columns = ['Generes','Overview','Keywords','Director','Top 5 Casts','Writer'])
```

```python
movies_new.reset_index(drop=True, inplace=True)
```

```python
movies_new['tags'] = movies_new['tags'].apply(lambda x: " ".join(x))
```

```python
## function to get stem/root words from tags
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
def stem(text):
    ps = PorterStemmer()
    x =[]
    for i in text.split():
        x.append(ps.stem(i))
    
    return " ".join(x)
```

```python
movies_new["tags"] = movies_new["tags"].apply(stem)
```

```python
movies_new
```

## Model Creation

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=15000,stop_words='english')
```

```python
vector = cv.fit_transform(movies_new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
def get_index(movie_name):
    idx = movies_new[movies_new['movie title'] == movie_name].index[0]
    return idx
```

```python
def recommend_1(movie):
    movie = movie[0]
    index = movies_new[movies_new['movie title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(movies_new.iloc[i[0]]['movie title'])
```

```python
def recommend_2(movie_list):
    index1= get_index(movie_list[0])
    index2= get_index(movie_list[1])
    pd1 = pd.DataFrame(list(enumerate(similarity[index1])),columns=['index number','distance']) 
    pd2 = pd.DataFrame(list(enumerate(similarity[index2])),columns=['index number','distance']) 
    pd4=pd.merge(pd1, pd2, on='index number', how='outer')
    pd4['average_dist'] = (pd4['distance_y'] + pd4['distance_x'])/2
    pd4 = pd4.sort_values(by = 'average_dist',ascending = False)
    for i in pd4['index number'][2:12].index:
        print(movies_new.iloc[i]['movie title'])
```

## We will be taking the cosine distance of all movies with 3 movies selected by user, Find the top movies with highest average cosine distance to the 3 movies selected and print them

```python
def recommend_3(movie_list):
    index1= get_index(movie_list[0])
    index2= get_index(movie_list[1])
    index3= get_index(movie_list[2])
    pd1 = pd.DataFrame(list(enumerate(similarity[index1])),columns=['index number','distance']) 
    pd2 = pd.DataFrame(list(enumerate(similarity[index2])),columns=['index number','distance']) 
    pd3 = pd.DataFrame(list(enumerate(similarity[index3])),columns=['index number','distance']) 
    pd4=pd.merge(pd1, pd2, on='index number', how='outer').merge(pd3, on='index number', how='outer')
    pd4['average_dist'] = (pd4['distance'] + pd4['distance_x'] + pd4['distance_y'])/3
    pd4 = pd4.sort_values(by = 'average_dist',ascending = False)
    for i in pd4['index number'][3:14].index:
        print(movies_new.iloc[i]['movie title'])
```

## Function to decide if we are inputing 1,2 or 3 movies

```python
def recommend(movie_list):
    if len(movie_list) == 1:
        return recommend_1(movie_list)
    elif len(movie_list) == 2:
        return recommend_2(movie_list)
    elif len(movie_list) == 3:
        return recommend_3(movie_list)
```

## Final Model

```python
movies = ['Kill Bill: Vol. 1']
```

```python
recommend(movies)
```
